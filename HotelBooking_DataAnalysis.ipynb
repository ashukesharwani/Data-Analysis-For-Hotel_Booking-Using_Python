# Data Analysis and Visualization

```python
# Analysis is performed by keeping in mind the problem statement
```

```python
# 1. check the amount of reservation which is canceled and which are not canceled

cancelled_percentages =df['is_canceled'].value_counts(normalize=True) 
# values_counts is the function which return the category name and number of times present
# and to return the in the percentage form we will give "normalised = True"
print(cancelled_percentages)

plt.figure(figsize=(5,4))
plt.title("Reservation status count")
plt.bar(["Not Canceled","Canceled"],df['is_canceled'].value_counts(),edgecolor='k',width=0.7)
plt.show()
```

```python
# 2. Now we need to check which hotel has higest cancelation ration

plt.figure(figsize=(8,4))
ax1=sns.countplot(x ='hotel',hue='is_canceled',data=df,palette='Blues')
lagend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title("Reservation status in different hotel",size=20)
plt.xlabel('hotel')
plt.ylabel('number of reservation')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python
#3. check the percentage of cancelation for both the hotel

#This is for the Resort hotel
resort_hotel=df[df['hotel']=='Resort Hotel'] #filter the data
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
#This is for the City hotel
city_hotel=df[df['hotel']=='City Hotel'] #filter the data
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
#4 Is there is affect of price in the cancellation of hotel
# Here we are gouping based on the reservation_data and mean price 
resort_hotel_group = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel_group=city_hotel.groupby('reservation_status_date')[['adr']].mean()
resort_hotel_group
```

```python
plt.figure(figsize=(20,8))
plt.title("Average Daily Rate on City and Resort Hotel",fontsize=30)
plt.plot(resort_hotel_group.index,resort_hotel_group['adr'],label="Resort Hotel")
plt.plot(city_hotel_group.index,city_hotel_group['adr'],label="City Hotel")
plt.legend(fontsize=20)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a list of months in the desired order
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']

plt.figure(figsize=(16, 8))
ax1 = sns.countplot(x=df['month'], hue='is_canceled', data=df, palette='bright', order=month_order)
legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1, 1))
plt.title("Reservation status per month", size=20)
plt.xlabel('Month')
plt.ylabel('Number of Reservations')
plt.legend(['Not Canceled', 'Canceled'])
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a list of months in the desired order
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']

plt.figure(figsize=(15, 8))
plt.title("ADR per month", fontsize=30)
df['month_year'] = df['month']
sns.barplot(x='month_year', y='adr', data=df[df['is_canceled'] == 1].groupby('month_year')[['adr']].sum().reset_index(), order=month_order)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.xlabel('Month')
plt.ylabel('ADR')
plt.show()
```

```python
#7 Let check the cancelation rate based on the countries
# we will see just the top 10 countries
cancelled_data=df[df['is_canceled'] == 1]
not_cancelled_data=df[df['is_canceled'] == 0]
top_10_countries = cancelled_data['country'].value_counts()[:10]
top_10_countries
```

```python
#from the graph we can find that PRT country has the higest cancellation 
import matplotlib.pyplot as plt

# Set up the data for the pie chart
top_10_countries = top_10_countries.head(10)  # Limit to top 10 countries

# Create a color palette for the pie chart
colors = plt.cm.Set3(range(len(top_10_countries)))

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.title("Top 10 Countries with Reservation Cancellations", fontsize=16)
plt.pie(top_10_countries, autopct='%.2f', labels=top_10_countries.index, colors=colors, startangle=90, wedgeprops={'edgecolor': 'white'}, pctdistance=0.85, labeldistance=1.05, shadow=True)

# Customize the legend
plt.legend(labels=top_10_countries.index, loc='best', fontsize=10)

# Add a circle at the center of the pie chart to make it look like a donut chart
center_circle = plt.Circle((0, 0), 0.60, fc='white')
plt.gca().add_artist(center_circle)

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Display the chart
plt.show()
# The data is in percentages
```

```python
#10 let's check the price whethere the hotel which is cancelled is heigh or the hotel which is not canceld is height
cancelled_df_adr=cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values("reservation_status_date",inplace=True)

not_cancelled_df_adr= not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values("reservation_status_date",inplace=True)

plt.figure(figsize=(20,6))
plt.title("Average Daily Rate ",fontsize=30)
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'],label='not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'],label='cancelled')
plt.legend(fontsize=20)
plt.show()
```
